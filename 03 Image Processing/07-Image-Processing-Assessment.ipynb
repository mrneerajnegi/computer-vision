```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
def display_img(img,cmap=None):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap)
```

**TASK: Open and display the giaraffes.jpg image that is located in the DATA folder.**

```python
giraf_image=cv2.imread("../Computer-Vision-with-Python/DATA/giraffes.jpg")
giraf_image=cv2.cvtColor(giraf_image,cv2.COLOR_BGR2RGB)
display_img(giraf_image)
```

**TASK:Apply a binary threshold onto the image.**

```python
giraf_image=cv2.imread("../Computer-Vision-with-Python/DATA/giraffes.jpg",0)
threshold,image=cv2.threshold(giraf_image,100,255,cv2.THRESH_BINARY)
image.shape
display_img(image,"gray")
```

**TASK: Open the giaraffes.jpg file from the DATA folder and convert its colorspace to  HSV and display the image.**

```python
girafImage=cv2.imread("../Computer-Vision-with-Python/DATA/giraffes.jpg")
hsv_image=cv2.cvtColor(girafImage,cv2.COLOR_BGR2HSV)
display_img(hsv_image)
```

**TASK: Create a low pass filter with a 4 by 4 Kernel filled with values of 1/10 (0.01) and then use 2-D Convolution to blur the giraffer image (displayed in normal RGB)**

```python
kernel=np.ones((4,4))/10
kernel
```

```python
display_img(cv2.filter2D(giraf_image,-1,kernel))
```

**TASK: Create a Horizontal Sobel Filter (sobelx from our lecture) with a kernel size of 5 to the grayscale version of the giaraffes image and then display the resulting gradient filtered version of the image.**
