```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
def loadImage():
    img=np.zeros((600,600))
    fontFace=cv2.FONT_HERSHEY_SIMPLEX
    cv2.putText(img,"NEERAJ",org=(10,400),fontFace=fontFace,fontScale=5,color=(255,255,255),thickness=10)
    return img;
def showImage(img):
    plt.figure(figsize=(15,8))
    plt.imshow(img,cmap="gray");
```

```python
img=loadImage()
showImage(loadImage())
```

```python
kernel=np.ones((6,6),dtype=np.uint8)
kernel
```

```python
eroded_img=cv2.erode(img,kernel,iterations=1)
showImage(eroded_img)
```

```python
#backgroundNoise
whiteNoisy=np.random.randint(low=0,high=2,size=(600,600))*255
showImage(whiteNoisy)
```

```python
noisyImage=img+whiteNoisy
showImage(noisyImage)
```

```python
#background noise
without_noise=cv2.morphologyEx(noisyImage,kernel=kernel,op=cv2.MORPH_OPEN)
showImage(without_noise)
```

```python
#foreground noise
img2=loadImage()
randomImage=np.random.randint(0,2,(600,600))*-255
img3=img2+randomImage
img3[img3==-255]=0
showImage(img3)
```

```python
withoutNoise=cv2.morphologyEx(img3,cv2.MORPH_CLOSE,kernel=kernel)
```

```python
showImage(withoutNoise)
```

```python
img=loadImage()
gradientImage=cv2.morphologyEx(img,cv2.MORPH_GRADIENT,kernel)
showImage(gradientImage)
```
