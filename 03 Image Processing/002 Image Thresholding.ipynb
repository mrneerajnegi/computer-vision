```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
image=cv2.imread("../Computer-Vision-with-Python/Data/rainbow.jpg",0) # 0 => gray image
plt.imshow(image,cmap="gray")
```

```python
def showImage(img):
    plt.figure(figsize=(15,12))
    ax=plt.subplot(111)
    ax.imshow(img,cmap="gray")
    
```

```python
showImage(image)
```

```python
threshold,thres_image=cv2.threshold(image,127,255,cv2.THRESH_BINARY)
showImage(thres_image)
```

```python
threshold,thres_image=cv2.threshold(image,127,255,cv2.THRESH_BINARY_INV)
showImage(thres_image)
```

```python
image=cv2.imread("../Computer-Vision-with-Python/Data/crossword.jpg",0) # 0 => gray image
showImage(image)
```

```python
threshold,threshimage=cv2.threshold(image,160,255,cv2.THRESH_BINARY)
showImage(threshimage)
```

```python
adaptiveImage=cv2.adaptiveThreshold(image,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,11,12) #AUTOMATIC THRESHOLD
showImage(adaptiveImage)
```

```python
image=cv2.addWeighted(threshimage,0.6,adaptiveImage,0.4,0)
showImage(image)
```
