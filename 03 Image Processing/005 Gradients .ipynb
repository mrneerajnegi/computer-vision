```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
def loadImage():
    image=cv2.imread("../Computer-Vision-with-Python/DATA/sudoku.jpg",0) #rgb
    return image;
def showImage(img):
    plt.figure(figsize=(15,10))
    plt.imshow(img,cmap="gray")
```

```python
showImage(loadImage())
```

```python
image=loadImage()
x_gradient=cv2.Sobel(image,cv2.CV_64F,1,0,ksize=5)
showImage(x_gradient)
```

```python
image=loadImage()
y_gradient=cv2.Sobel(image,cv2.CV_64F,0,1,ksize=5)
showImage(y_gradient)
```

```python
image=loadImage()
laplacian=cv2.Laplacian(image,cv2.CV_64F,ksize=5)
showImage(laplacian)
```

```python
mixImage=cv2.addWeighted(src1=x_gradient,alpha=0.5,src2=y_gradient,beta=0.6,gamma=0)
showImage(mixImage)
```

```python
kernel=np.ones((5,5))
```

```python
image=loadImage()
mixImage=cv2.morphologyEx(image,cv2.MORPH_GRADIENT,kernel=kernel)
showImage(mixImage)
```
