```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
doggy_img=cv2.imread("../Computer-Vision-with-Python/DATA/dog_backpack.jpg")
dnc_img=cv2.imread("../Computer-Vision-with-Python/DATA/watermark_no_copy.png")
doggy_img=cv2.cvtColor(doggy_img,cv2.COLOR_BGR2RGB)
dnc_img=cv2.cvtColor(dnc_img,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(doggy_img)
```

```python
plt.imshow(dnc_img)
```

```python
#if size are same then it is so easy to blend the images
doggy_img=cv2.resize(doggy_img,(1200,900))
dnc_img=cv2.resize(dnc_img,(1200,900))
```

```python
plt.imshow(doggy_img)
```

```python
blended_image=cv2.addWeighted(src1=doggy_img,alpha=1,src2=dnc_img,beta=0.1,gamma=0)
```

```python
plt.imshow(blended_image)
```

```python
# blending different size images can be done by numpy
doggy_img=cv2.imread("../Computer-Vision-with-Python/DATA/dog_backpack.jpg")
dnc_img=cv2.imread("../Computer-Vision-with-Python/DATA/watermark_no_copy.png")
doggy_img=cv2.cvtColor(doggy_img,cv2.COLOR_BGR2RGB)
dnc_img=cv2.cvtColor(dnc_img,cv2.COLOR_BGR2RGB)
```

```python
print(doggy_img.shape,dnc_img.shape)
```

```python
dnc_img=cv2.resize(dnc_img,(600,600))
```

```python
x_offset=0
y_offset=0
doggy_img[x_offset:y_offset+dnc_img.shape[1],y_offset:y_offset+dnc_img.shape[0],:]=dnc_img
```
