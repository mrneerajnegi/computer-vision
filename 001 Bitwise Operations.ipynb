```python
import numpy as np
import cv2
import matplotlib.pyplot  as plt
```

```python
image=cv2.imread("./Computer-Vision-with-Python/DATA/00-puppy.jpg",0)
plt.imshow(image,cmap="gray")
```

```python
def showImgae(image):
    plt.figure(figsize=(12,8))
    plt.imshow(image,cmap="gray")
```

```python
mask=np.zeros(image.shape,dtype=np.uint8)
cv2.circle(mask,(1500,800),100,255,-1)
showImgae(mask)
```

```python
mask1=np.zeros(image.shape)
cv2.rectangle(mask1,(1100,400),(1500,800),255,-1)
showImgae(mask1)
```

```python
aa=cv2.bitwise_and(mask,mask1)
showImgae(aa)
```

```python
bb=cv2.bitwise_or(mask,mask1)
showImgae(bb)
```

```python
cc=cv2.bitwise_not(mask,mask1)
showImgae(cc)
```

```python
dd=cv2.bitwise_and(mask1,mask1)
showImgae(dd)
```

```python
dd=cv2.bitwise_or(mask1,mask1,mask=image)
showImgae(dd)
```

```python
import numpy as np
import cv2
import matplotlib.pyplot  as plt
background=None
def cal_acc(frame):
    global background
    if background is None:
        background=frame.copy().astype("float")
    averageValue1 = np.float32(frame)
    cv2.accumulateWeighted(averageValue1,background,0.5)
```

```python
cap=cv2.VideoCapture(0)
while True:
    ret ,frame=cap.read()
    frameGray=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    frameGray=cv2.GaussianBlur(frameGray,(7,7),0)
#     thresh,frame=cv2.threshold(frame,100,255,cv2.THRESH_BINARY)
    cal_acc(frameGray)
    cv2.imshow("frame",frame)
    if cv2.waitKey(1) & 0xFF==27:
        break
cap.release()
cv2.destroyAllWindows()
```
