```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
chessBoard=cv2.imread("../Computer-Vision-with-Python/DATA/flat_chessboard.png")
circleBoard=cv2.imread("../Computer-Vision-with-Python/DATA/dot_grid.png")
plt.subplot(121)
plt.imshow(chessBoard)
plt.subplot(122)
plt.imshow(circleBoard)
plt.show()
```

```python
found,corners=cv2.findChessboardCorners(chessBoard,(5,5))
```

```python
found
```

```python
copy=chessBoard.copy()
cv2.drawChessboardCorners(copy,(5,5),corners,found)
plt.figure(figsize=(10,8))
plt.imshow(copy)
```

```python
found,corners=cv2.findCirclesGrid(chessBoard,(5,5))
```

```python
found,corners=cv2.findCirclesGrid(circleBoard,(5,5))
```
