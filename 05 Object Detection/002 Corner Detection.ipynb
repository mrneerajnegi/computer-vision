```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
chess=cv2.imread("../Computer-Vision-with-Python/DATA/flat_chessboard.png")
grayChess=cv2.cvtColor(chess,cv2.COLOR_BGR2GRAY)
chess=cv2.cvtColor(chess,cv2.COLOR_BGR2RGB)
realChess=cv2.imread("../Computer-Vision-with-Python/DATA/real_chessboard.jpg")
grayRealChess=cv2.cvtColor(realChess,cv2.COLOR_BGR2GRAY)
realChess=cv2.cvtColor(realChess,cv2.COLOR_BGR2RGB)
```

```python
plt.figure(figsize=(15,12))
plt.subplot(121)
plt.imshow(chess)
plt.subplot(122)
plt.imshow(realChess)
plt.show()
```

```python
plt.figure(figsize=(15,12))
plt.subplot(121)
plt.imshow(grayChess,cmap="gray")
plt.subplot(122)
plt.imshow(grayRealChess,cmap="gray")
plt.show()
```

# CornerHarris

```python
#corners detect on grayscale image only
corners=cv2.cornerHarris(grayChess,2,3,0.04)
dist=cv2.dilate(corners,None)
```

```python
chess[dist>0.01*dist.max()]=[255,0,0]
plt.figure(figsize=(10,10))
plt.imshow(chess)
```

```python
corner=cv2.cornerHarris(grayRealChess,3,3,0.01)
dist=cv2.dilate(corner,None)
```

```python
realChess[dist>0.01*dist.max()]=[255,0,0]
plt.figure(figsize=(10,10))
plt.imshow(realChess)
```

# Shi-Tomasi Detection

```python
chess=cv2.imread("../Computer-Vision-with-Python/DATA/flat_chessboard.png")
grayChess=cv2.cvtColor(chess,cv2.COLOR_BGR2GRAY)
chess=cv2.cvtColor(chess,cv2.COLOR_BGR2RGB)
realChess=cv2.imread("../Computer-Vision-with-Python/DATA/real_chessboard.jpg")
grayRealChess=cv2.cvtColor(realChess,cv2.COLOR_BGR2GRAY)
realChess=cv2.cvtColor(realChess,cv2.COLOR_BGR2RGB)
```

```python
corner=cv2.goodFeaturesToTrack(grayChess,150,0.01,10)
```

```python
corner=np.int0(corner)
for i in corner:
    x,y=i.ravel()
    cv2.circle(chess,(x,y),1,(255,0,0),10)
```

```python
plt.figure(figsize=(10,10))
plt.imshow(chess)
```
