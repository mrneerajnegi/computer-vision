# Always find edges after bluring the image

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
dog=cv2.imread("../Computer-Vision-with-Python/DATA/sammy_face.jpg")
```

```python
plt.figure(figsize=(15,10))
plt.imshow(dog)
```

```python
#formula to finding the thresholds :
pixedMedian=np.median(dog)   
lower=int(max(0,0.7*pixedMedian))
higher=int(min(255,1.3*pixedMedian))
```

```python
plt.figure(figsize=(15,10))
edgeImages=cv2.Canny(dog,lower,higher)
plt.imshow(edgeImages)
```

```python
blurredImage=cv2.blur(dog,(5,5))
plt.figure(figsize=(15,10))
edgeImages=cv2.Canny(blurredImage,lower,higher)
plt.imshow(edgeImages)
```

```python
#random thresholds
plt.figure(figsize=(15,10))
edgeImages=cv2.Canny(dog,200,300)
plt.imshow(edgeImages)
```
