```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
```

```python
blank_image=np.zeros(shape=(512,512,3),dtype=np.int32)
```

```python
plt.imshow(blank_image)
```

```python
cv2.rectangle(blank_image,pt1=(400,10),pt2=(500,100),color=(0,0,255),thickness=10)
plt.imshow(blank_image)
```

```python
cv2.rectangle(blank_image,pt1=(400,140),pt2=(500,260),color=(0,0,255),thickness=-1)
plt.imshow(blank_image)
```

```python
cv2.circle(blank_image,center=(250,250),radius=50,color=(0,255,0),thickness=10)
plt.imshow(blank_image)
```

```python
cv2.line(blank_image,pt1=(0,0),pt2=(512,512),color=(255,0,0),thickness=10)
plt.imshow(blank_image)
```

```python
font=cv2.FONT_ITALIC
cv2.putText(blank_image,text="Hello",fontFace=font,color=(255,255,255),fontScale=1,org=(20,500),
            thickness=3,lineType=cv2.LINE_AA)
plt.imshow(blank_image)
```

```python
vetices=np.array([[50,400],[200,200],[400,100],[450,400]])
```

```python
#we need 3 dimentions for polygon
vetices.shape
```

```python
vetices=vetices.reshape(-1,1,2)
```

```python
vetices.shape
```

```python
vetices
```

```python
blank_image=np.zeros(shape=(512,512,3),dtype=np.int32)
```

```python
cv2.polylines(blank_image,[vetices],isClosed=False,color=(0,255,255),thickness=5)
plt.imshow(blank_image)
```

```python
blank_image=np.zeros(shape=(512,512,3),dtype=np.int32)
cv2.polylines(blank_image,[vetices],isClosed=True,color=(0,255,255),thickness=5)
plt.imshow(blank_image)
```
